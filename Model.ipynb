```python
import os
import librosa
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from tqdm import tqdm

# âœ… Check for GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
# âœ… Dataset CSV File (Ensure this file exists)
CSV_FILE = "asvspoof2017_labels.csv"
AUDIO_DIR = "./ASVspoof2017_V2_train/ASVspoof2017_V2_train/"
SAMPLE_RATE = 16000
N_MFCC = 13
MAX_LEN = 400  # Maximum time frames for padding
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()  # Standardize MFCC features

def extract_mfcc(file_path, n_mfcc=N_MFCC, max_len=MAX_LEN):
    y, sr = librosa.load(file_path, sr=SAMPLE_RATE)
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)

    if mfcc.shape[1] < max_len:
        pad_width = max_len - mfcc.shape[1]
        mfcc = np.pad(mfcc, ((0, 0), (0, pad_width)), mode='constant')
    else:
        mfcc = mfcc[:, :max_len]

    # âœ… Normalize MFCC (Standardization)
    mfcc = scaler.fit_transform(mfcc)  # Apply Standard Scaling
    return mfcc
```

```python
df=pd.read_csv(CSV_FILE)
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['label'])
label_encoder = LabelEncoder()
train_df['label'] = label_encoder.fit_transform(train_df['label'])  
test_df['label'] = label_encoder.transform(test_df['label'])  # Only transform

# âœ… Custom Dataset Class
class AudioDataset(Dataset):
    def __init__(self, data, audio_dir, max_len=MAX_LEN):
        self.data = data.reset_index(drop=True)
        self.audio_dir = audio_dir
        self.max_len = max_len

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        file_name = self.data.iloc[idx, 0]  # First column: Filename
        label = self.data.iloc[idx, 1]  # Encoded Label
        file_path = os.path.join(self.audio_dir, file_name)
        mfcc_features = extract_mfcc(file_path, max_len=self.max_len)

        return torch.tensor(mfcc_features, dtype=torch.float32), torch.tensor(label, dtype=torch.long)

# âœ… Create Train & Test Datasets
train_dataset = AudioDataset(train_df, AUDIO_DIR)
test_dataset = AudioDataset(test_df, AUDIO_DIR)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

```

```python
import torch
import torch.nn as nn

class CNN_BiLSTM(nn.Module):
    def __init__(self, input_dim=13, hidden_dim=128, num_layers=2, num_classes=2):
        super(CNN_BiLSTM, self).__init__()

        # âœ… CNN Layers
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=32, kernel_size=(3, 3), padding=1)
        self.bn1 = nn.BatchNorm2d(32)  # âœ… BatchNorm for stability
        self.conv2 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=(3, 3), padding=1)
        self.bn2 = nn.BatchNorm2d(64)
        self.relu = nn.ReLU()

        # âœ… Adaptive Pooling
        self.pool = nn.AdaptiveAvgPool2d((input_dim, 32))  # Reducing feature size for LSTM

        # âœ… BiLSTM
        self.lstm = nn.LSTM(input_size=32, hidden_size=hidden_dim, num_layers=num_layers,
                            bidirectional=True, batch_first=True, dropout=0.3)

        # âœ… Fully Connected Layer
        self.fc = nn.Linear(hidden_dim * 2, num_classes)
        self.dropout = nn.Dropout(0.3)

    def forward(self, x):
        # âœ… Ensure the input is 4D (batch, channels, height, width)
        x = x.unsqueeze(1) if x.dim() == 3 else x  

        # âœ… CNN Feature Extraction
        x = self.relu(self.bn1(self.conv1(x)))  
        x = self.relu(self.bn2(self.conv2(x)))  

        # âœ… Apply Pooling
        x = self.pool(x)  

        # âœ… Reshape for LSTM
        batch_size, channels, height, width = x.shape  
        x = x.permute(0, 2, 1, 3).contiguous()  # (batch, height, channels, width)
        x = x.view(batch_size, height, width)  # (batch_size, time_steps, features)

        # âœ… Pass through LSTM
        x, _ = self.lstm(x)

        # âœ… Fully Connected Layer with Dropout
        x = self.fc(self.dropout(x[:, -1, :]))

        return x
```

```python
model = CNN_BiLSTM().to(device)
weights = torch.tensor([1.0, 3.0]).to(device)  # Adjust weights based on class imbalance
criterion = nn.CrossEntropyLoss(weight=weights)

optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-5)  # âœ… L2 Regularization
```

```python
def train_model(model, train_loader, epochs):
    model.train()
    for epoch in range(epochs):
       for inputs, labels in train_loader:
            print("Before passing to model:", inputs.shape)  # Debugging

            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            
            outputs = model(inputs)  # <-- This line is causing the error
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
```

```python
# âœ… Evaluation Function
def evaluate_model(model, test_loader):
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            predictions = torch.argmax(outputs, dim=1)
            correct += (predictions == labels).sum().item()
            total += labels.size(0)
    print(f"Test Accuracy: {correct/total:.4f}")
```

```python
# âœ… Train & Evaluate
train_model(model, train_loader, epochs=5)
evaluate_model(model, test_loader)

# âœ… Save Model
torch.save(model.state_dict(), "hybrid_audio_detector.pth")
print("Model saved successfully!")
```

```python
# âœ… Define Test Dataset Paths
audio_dir = "./ASVspoof2017_V2_eval/ASVspoof2017_V2_eval/"
csv_file = "asvspoof2017_labelseval.csv"

# âœ… Load CSV File & Apply Label Encoding
test_df = pd.read_csv(csv_file)
test_df['label'] = label_encoder.transform(test_df['label'])  # Use previously fitted encoder

# âœ… Initialize the Test Dataset & DataLoader
test_dataset = AudioDataset(test_df, audio_dir)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# âœ… Verify Dataset Loading
print(f"âœ… Test Dataset Loaded: {len(test_dataset)} samples")
```

```python
import torch
import torch.nn.functional as F

def test_model(model, test_loader, device):
    model.eval()  # âœ… Put model in evaluation mode
    correct = 0
    total = 0

    with torch.no_grad():  # âœ… No gradient updates needed
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)  # âœ… Move data to correct device

            # âœ… Forward pass
            outputs = model(inputs)  

            # âœ… Convert outputs to probabilities
            probs = F.softmax(outputs, dim=1)

            # âœ… Get predicted class index
            _, predicted = torch.max(probs, dim=1)

            # âœ… Compute Accuracy
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    # âœ… Final Accuracy Calculation
    accuracy = 100 * correct / total
    print(f"\nâœ… Test Accuracy: {accuracy:.2f}%")
    return accuracy  # Return for logging if needed

# âœ… Run Test
test_model(model, test_loader, device)

```

```python
import torch

def load_model(model, checkpoint_path, device):
    model.load_state_dict(torch.load(checkpoint_path, map_location=device))
    model.to(device)
    print("âœ… Model Loaded Successfully!")
    return model

```

```python
def fine_tune_model(model, train_loader, epochs, optimizer, criterion, device):
    model.train()  # âœ… Training Mode

    for epoch in range(epochs):
        correct, total, running_loss = 0, 0, 0.0

        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(inputs)

            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            total += labels.size(0)
            _, predicted = torch.max(outputs, 1)
            correct += (predicted == labels).sum().item()

        accuracy = 100 * correct / total
        print(f"ðŸ”¥ Epoch {epoch+1}: Loss={running_loss:.4f}, Accuracy={accuracy:.2f}%")

    print("\nâœ… Fine-tuning Complete!")
```

```python
# âœ… Load Model
checkpoint_path = "hybrid_audio_detector.pth"  # Change this to your actual saved model path
model = load_model(model, checkpoint_path, device)

# âœ… Define Optimizer & Loss Again (for fine-tuning)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-5)  # Lower LR for fine-tuning
criterion = torch.nn.CrossEntropyLoss()

# âœ… Fine-Tune Model on Existing Data
fine_tune_model(model, train_loader, epochs=5, optimizer=optimizer, criterion=criterion, device=device)

# âœ… Evaluate Again
test_model(model, test_loader, device)
```
